# Setting up the Data

Let's set up our data for our Random Forest Classification Model

```python
loans.info()
```

## Categorical Features

First I'll transform the "purpose" column using dummy variables so that I can use sklearn (This will create a new feature for each possible purpose, and a simple True/1 or False/0).

```python
cat_feats = ['purpose']
```

```python
pd.get_dummies(loans,columns=cat_feats,drop_first=True).head()
```

```python
final_data = pd.get_dummies(loans,columns=cat_feats,drop_first=True)
```

## Train Test Split

Now its time to split our data into a training set and a testing set!

```python
from sklearn.model_selection import train_test_split
```

```python
X = final_data.drop(['not.fully.paid'],axis=1)
y = final_data['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

## Training a Decision Tree Model

Let's start by training a single decision tree first!

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Predictions and Evaluation of Decision Tree

```python
pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print (classification_report(y_test,pred))
print ('\n')
print (confusion_matrix(y_test,pred))
```

## Training the Random Forest model

Now its time to train our Random Forest model!

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

## Predictions and Evaluation

```python
rfc_pred = rfc.predict(X_test)
```

```python
print (classification_report(y_test,rfc_pred))
```

```python
print (confusion_matrix(y_test,rfc_pred))
```

Comparing the two methods, looks like Random Forest is the better fit for this question. Our precision (of all lendees we predicted to pay in full, how many did?) is 78% compared to 75%, while recall (of all lendees who paid their loan, how many did we predict?) is 84% vs. 74%.
